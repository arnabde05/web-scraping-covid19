```python
import requests
from bs4 import BeautifulSoup
```

```python
# Send a GET request to the website
url = 'https://www.ndtv.com/coronavirus/fullcoverage#statesmap'
response = requests.get(url)
```

```python
# Create BeautifulSoup object and specify the parser
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
# Find the main container containing the COVID-19 data
container = soup.find('div', {'class': 'site-mid-part'})
```

```python
# Find the table containing the COVID-19 data within the container
table = container.find('table')

# Extract the table rows
rows = table.find_all('tr')
```

```python
# Iterate over the rows and extract the data
for row in rows:
    cells = row.find_all('td')
    if len(cells) == 5:  # Ensure it's a valid row
        state = cells[0].text.strip()
        affected = cells[1].text.strip()
        death = cells[2].text.strip()
        recovered = cells[3].text.strip()
        active = cells[4].text.strip()

        print('State:', state)
        print('Affected:', affected)
        print('Death:', death)
        print('Recovered:', recovered)
        print('Active:', active)
        print('---')
```
