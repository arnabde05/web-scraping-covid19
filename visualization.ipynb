```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
covid19 = pd.read_csv("cleaned_covid19.csv")
covid_data_prs = pd.read_csv("covid_data_prs.csv")
mohfw = pd.read_csv("cleaned_mohfw.csv")
```

```python
covid19.info()
```

```python
variables = ["affected", "death", "recovered","active"]
data = {
    'covid19': covid19[variables].sum(),
    'PRS': covid_data_prs[variables].sum(),
    'MOHFW': mohfw[variables].sum()
}
```

```python
# Create the bar chart
plt.figure(figsize=(10, 6))

# Define the bar width
bar_width = 0.2

# Generate evenly spaced values for x-axis ticks
x = np.arange(len(variables))

# Offset for each website's bars
offsets = [-bar_width, 0, bar_width]

# Iterate over the websites and plot the bars
for i, (website, values) in enumerate(data.items()):
    # Shift the x-values for each website to avoid overlapping bars
    plt.bar(x + offsets[i], values, width=bar_width, label=website)

# Customize the chart
plt.title('COVID-19 Division State-wise Data in India')
plt.xlabel('Variables')
plt.ylabel('Count')
plt.legend()

# Set the x-axis tick labels
plt.xticks(x, variables)

# Adjust the layout
plt.tight_layout()

# Display the chart
plt.show()
```
