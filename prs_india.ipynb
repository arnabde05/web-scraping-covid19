```python
import requests
from bs4 import BeautifulSoup

# Send a GET request to the webpage
url = "https://prsindia.org/covid-19/cases"
response = requests.get(url)

# Create a BeautifulSoup object
soup = BeautifulSoup(response.text, 'html.parser')

# Find the table containing the COVID-19 data
table = soup.find('table', class_='table table-striped table-bordered')

# Initialize lists to store the scraped data
states = []
affected_cases = []
active_cases = []
recovered_cases = []
death_cases = []

# Find all the rows in the table body
rows = table.tbody.find_all('tr')

# Iterate over the rows and extract the data
for row in rows:
    # Extract the data from each column in the row
    columns = row.find_all('td')
    state = columns[1].text.strip()
    affected = columns[2].text.strip()
    active = columns[3].text.strip()
    recovered = columns[4].text.strip()
    death = columns[5].text.strip()

    # Append the data to the respective lists
    states.append(state)
    affected_cases.append(affected)
    active_cases.append(active)
    recovered_cases.append(recovered)
    death_cases.append(death)

# Print the scraped data
for i in range(len(states)):
    print(f"State: {states[i]}")
    print(f"Affected: {affected_cases[i]}")
    print(f"Active: {active_cases[i]}")
    print(f"Recovered: {recovered_cases[i]}")
    print(f"Death: {death_cases[i]}")
    print("---------------------------")
```
